```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\cse\Desktop\insurance.csv")
```

```python
df.head()
```

```python
def map_smoking (column):
    mapped = []
    for i in column:
        if i == "yes":
            mapped.append(1)
        else:
            mapped.append(0)
    return mapped
```

```python
new_col = map_smoking(df["smoker"])
```

```python
new_col
```

```python
df["smoker"] = new_col
df.head()
```

```python
def map_sex(column):
    mapped = []
    for i in column:
        if i == "male":
            mapped.append(1)
        else:
            mapped.append(0)
    return mapped

df["sex"] = map_sex(df["sex"])
df.head()
```

```python

df.columns
```

```python
fig,axes = plt.subplots(nrows = 1 , ncols = 2, figsize = (12,4))
sns.countplot(x="sex", data = df, hue="smoker" ,ax=axes[0] )
sns.barplot(x="sex",y = "charges",data = df,hue = "smoker", ax = axes[1])
```

```python
sns.countplot(x = "children",data=df)
```

```python

sns.heatmap(df.corr(),annot = True)
```

```python
x = df[['age', 'bmi', 'children', 'smoker', 'sex']]
y = df[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
x_train.head()
```

```python

from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.intercept_
```

```python

lm.coef_
```

```python

pred = lm.predict(x_test)
```

```python
plt.scatter(y_test,pred)
```

```python
sns.distplot(y_test - pred)
```

```python
from sklearn.metrics import r2_score
```
